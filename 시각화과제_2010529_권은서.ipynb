# 데이터 조작하기


```python
# 데이터 조작

esdf_new = esdf.copy()    # 복사본 만들기, df_new = df 를 사용하는 경우 이름만 다를 뿐 어느 한 쪽을 수정하면 다른 한 쪽도 같이 수정됨  
esdf_new = esdf_new.rename (columns = {'성별코드': 'Gender'})  
esdf_new = esdf_new.rename (columns = {'연령대코드(5세단위)':'Age'})
esdf_new = esdf_new.rename (columns = {'혈청크레아티닌':'Creatinine'})
esdf_new = esdf_new.rename (columns = {'감마지티피':'GTP'})
esdf_new = esdf_new.rename (columns = {'시도코드':'Area'})
esdf_new = esdf_new.rename (columns = {'구강검진수검여부':'Oral_Health_Check'})
display(esdf_new)
```

# 차트 및 그래프 만들기


```python
#불러오기

%matplotlib inline
import matplotlib as mpl               # 기본 설정
import matplotlib.pyplot as plt        # 그래프 관련
import seaborn as sns
```

**1. 나이대별로 혈청크레아티닌의 수치는 어떠한가?**

**[그래프 해석 및 인사이트 도출]**

- 혈청크레아티닌의 수치는 높아질수록 콩팥 등의 신장에 문제가 발생한 것이고, 낮을수록 별다른 문제가 발생하지 않았음을 의미한다. 다만, 임신이나 근육량 감소로 인한 수치 감소도 발생할 수 있으므로, 다양한 상황을 고려할 필요가 있다.
- 번호 9번(40~45세 해당)부터 번호 18번(85세 이상 해당)까지 0.8에서 1.0에서 수치를 웃돌고 있다. 혈중 크레아티닌 수치의 정상범위는 대략 0.5에서 1.4로,대부분의 연령대 사이에서 비슷한 수치를 보인다는 점을 알아낼 수 있다.
- 그러나, 연령대가 증가할수록 그래프가 미세하게 증가하게 된다는 점에서, 연령이 늘어날수록 혈청크레아티닌의 수치가 높아질 가능성이 있음을 의미한다. 이 결과를 활용해 신장 문제 관리의 해결방안을 마련하면 질환자들에게 많은 도움이 될 수 있을 것이다.



```python
# 맷플롯립 방식 활용
import matplotlib.pyplot as plt 
Age_Creatinine = esdf_new.groupby('Age')['Creatinine'].mean()
Age_sorted = Age_Creatinine.sort_values()[:10]
print(Age_sorted)

Age_list = Age_sorted.index.tolist()
Age_Creat = Age_sorted.values.tolist()

plt.bar(Age_list, Age_Creat, align='center', color = 'g')  # x 축을 연령대로, y축을 혈청크레아티닌의 평균으로 설정하고 수치를 비교하기
plt.ylabel('Value of Creatinine')
plt.title('Age(ascending)')
plt.xticks(rotation=45) 
plt.show()
```

**2. 성별에 따른 감마지티피 수치는 어떠한가?**

**[그래프 해석 및 인사이트 도출]**
- 감마지티피의 정상 범위는 남성이 11에서 63IU/L, 여성이 8에서35 IU/L으로, 평균적으로는 대부분 정상범위에 속한다는 점을 알 수 있다.
- 그러나 상자 그림을 통해 이상치를 확인했을 때, 남성의 인원수가 여성보다 많음을 고려하여도 상대적으로 이상치의 비중이 더 큰 것을 확인할 수 있다. 즉, 간 질환에 노출될 수 있는 위험성이 여성보다 더 높을 수 있음을 의미한다. 이에 대한 원인을 밝힘과 동시에 해결방안을 마련하면 좋을 것으로 보인다.

```python
#집단별 평균표 만들기
#씨본 패키지 활용
esdf_new_GTP = esdf_new.groupby('Gender' , as_index = False).agg( meanGTP = ('GTP', 'mean'))
esdf_new_GTP
```

```python
sns.barplot(data = esdf_new_GTP , x = 'Gender', y = 'meanGTP')
```

```python
sns.boxplot(data = esdf_new, x = 'Gender', y = 'GTP')
```

**3. 지역별 구강검진 수검 여부는 어떠한가? (경상도 부근을 중심으로)**

**[그래프 해석 및 인사이트 도출]**

- 가장 점이 높은 수치에 위치한 번호는 31번으로, 이는 울산광역시에 해당한다.
- 반면, 점이 가장 낮은 수치에 위치한 번호는 47번으로, 이는 경상북도에 해당한다. 그다음으로는 대구광역시, 부산광역시, 경상남도가 낮은 수치에 있다.
- 이를 통해 경상도 부근 지역에서의 구강검진 수검 여부 평균은 타지역에 비해 낮은 편이지만, 이례적으로 울산광역시의 경우 전체적으로 가장 높은 평균을 차지하게 되었다. 이와 같은 결과가 도출된 원인에 대하여 조사한다면 유익한 정보를 얻을 수 있을 것이다.

```python
Area_count = esdf_new['Area'].value_counts()
print(Area_count)
```

```python
#높은 평균 산출
#맷플롯립 방식 활용

print('시도:', len(esdf_new['Area'].unique())) 
#esdf_area= esdf_new.groupby('Area')['Oral_Health_Check'].sum()
#esdf_area_sorted = esdf_area.sort_values(ascending=True)[9:17]
#print(esdf_area_sorted)

esdf_area= esdf_new.groupby('Area')['Oral_Health_Check'].mean()
esdf_area_sorted = esdf_area.sort_values(ascending = True)[9:17]
print(esdf_area_sorted)


Area_list = esdf_area_sorted.index.tolist()
Area_Oral_Health = esdf_area_sorted.values.tolist()

plt.scatter(Area_list, Area_Oral_Health, color="red", alpha=0.5, label="scatter point")
plt.xlabel("Area_List")
plt.ylabel("Oral_Health_Check")

plt.title('Regional Average: Number of people checked oral health ')
plt.xticks(rotation=45) 
plt.show()
```

```python
# 낮은 평균 산출

print('시도:', len(esdf_new['Area'].unique())) 
#esdf_area= esdf_new.groupby('Area')['Oral_Health_Check'].sum()
#esdf_area_sorted = esdf_area.sort_values(ascending=True)[1:8]
#print(esdf_area_sorted)

esdf_area= esdf_new.groupby('Area')['Oral_Health_Check'].mean()
esdf_area_sorted = esdf_area.sort_values(ascending=True)[1:8]
print(esdf_area_sorted)

Area_list = esdf_area_sorted.index.tolist()
Area_Oral_Health = esdf_area_sorted.values.tolist()

plt.scatter(Area_list, Area_Oral_Health, color="red", alpha=0.5, label="scatter point")
plt.xlabel("Area_List")
plt.ylabel("Oral_Health_Check")
plt.title('Number of people checked oral health (Regional Average)')
plt.xticks(rotation=45) 
plt.show()
```
